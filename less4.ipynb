```python
import requests
from lxml import html
import datetime
import pandas
from pymongo import MongoClient
```

```python
news_data = []
```

```python
u_a = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:69.0) Gecko/20100101 Firefox/69.0','Accept':'*/*'}

def pars_lenta():
    u_a = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:69.0) Gecko/20100101 Firefox/69.0',
           'Accept':'*/*'}
    source = 'https://www.lenta.ru'
    res = requests.get(source, headers = u_a)
    page = html.fromstring(res.text)
    news_page = page.xpath("//div[@class='span4']")

    for n_p in news_page:
        news_list = n_p.xpath(".//div[@class='item']|.//div[@class='first-item']")
        
        for news in news_list:
            news_data.append({'source':source,
                          'url':source+news.xpath('.//a/@href')[0],
                          'news_title':news.xpath('.//a/text()')[0].replace('\xa0',' '),
                          'date':news.xpath('.//a/time/@title')[0]})    
```

```python
def pars_yandex():
    u_a = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:69.0) Gecko/20100101 Firefox/69.0',
           'Accept':'*/*'}
    source = 'https://www.yandex.ru'
    res = requests.get(source, headers = u_a)
    page = html.fromstring(res.text)
    news_page = page.xpath("//ol")
    
    for n_p in news_page:
        news_list = n_p.xpath(".//li")
        
        for news in news_list:    
            news_data.append({'source':source,
                              'url':news.xpath('.//a/@href')[0],
                              'news_title':news.xpath('.//a/@aria-label')[0],
                              'date': datetime.date.today().strftime('%d %m %Y')})
        
```

```python
def pars_mail():
    
    u_a = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:69.0) Gecko/20100101 Firefox/69.0',
           'Accept':'*/*'}
    source = 'https://www.news.mail.ru'
    res = requests.get(source, headers = u_a)
    page = html.fromstring(res.text)
    news_page = page.xpath("//li")
    
    for news in news_page:    
        
        if len(news.xpath('.//a/text()')) == 0:
            news_title = None
        else:
            news_title = news.xpath('.//a/text()')[0]
        
        news_data.append({'source':source,
                              'url':source+news.xpath('.//a/@href')[0],
                              'news_title':news_title,
                              'date': datetime.date.today().strftime('%d %m %Y')})
```

```python
pars_lenta()
pars_yandex()
pars_mail()
```

```python
client = MongoClient('localhost', 27017)
db = client['news']
collection = db.news_col

for i in news_data:
    collection.insert_one(i)
    
```
